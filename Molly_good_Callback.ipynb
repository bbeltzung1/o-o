```python
import os
import matplotlib.pyplot as plt
import random
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image_dataset_from_directory

base_dir = 'C:/Users/bbeltzun/Downloads/Molly_split_subdir/'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'test')
```

```python

training_set = image_dataset_from_directory(train_dir,
                                             shuffle=True,
                                             batch_size=32,
                                             image_size=(224, 224),
                                             label_mode='categorical')
val_dataset = image_dataset_from_directory(validation_dir,
                                                shuffle=True,
                                                batch_size=32,
                                                image_size=(224, 224),
                                                label_mode='categorical')
```

```python
data_augmentation = keras.Sequential(
    [       keras.layers.experimental.preprocessing.RandomFlip("horizontal"),
     keras.layers.experimental.preprocessing.RandomFlip("vertical"),
    ]
)
```

```python
for images, labels in training_set.take(1):
    plt.figure(figsize=(12, 12))
    first_image = images[0]
    for i in range(12):
        ax = plt.subplot(3, 4, i + 1)
        augmented_image = data_augmentation(
            tf.expand_dims(first_image, 0)
        )
        plt.imshow(augmented_image[0].numpy().astype("int32"))
        plt.axis("off")
```

```python
def plot_model_history(model_history, accuracy='accuracy', val_accuracy='val_accuracy'):
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    axs[0].plot(range(1,len(model_history.history[accuracy])+1),model_history.history[accuracy])
    axs[0].plot(range(1,len(model_history.history[val_accuracy])+1),model_history.history[val_accuracy])
    #axs[0].set_xticks(np.arange(1,len(model_history.history[accuracy])+1),len(model_history.history[accuracy])/10)
    axs[0].legend(['train', 'val'], loc='best')
    axs[1].plot(range(1,len(model_history.history['loss'])+1),model_history.history['loss'])
    axs[1].plot(range(1,len(model_history.history['val_loss'])+1),model_history.history['val_loss'])
    axs[1].set_title('Model Loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    #axs[1].set_xticks(np.arange(1,len(model_history.history['loss'])+1),len(model_history.history['loss'])/10)
    axs[1].legend(['train', 'val'], loc='best')
    plt.savefig('C:/Users/bbeltzun/Documents/o-o/plots/Molly_good_callbacks/'+'batch'+str(batch)+
               '-oneFC'+str(fc_size)+'-Dropout'+str(dropout)+'-SGD'+str(lr)+'-unfreeze'+str(unfreeze)+
                '-l2reg'+str(l2_val)+'.png')
    #plt.show()
```

```python
epoch = 20

lrs = [0.1, 0.01, 0.001]
unfreezes = [0, 17, 'none']
l2_vals = [0.1, 0.01, 0.001]
fc_sizes = [2048]
batch_sizes = [16, 32]
dropouts = [0.2,0.4]


i=0

for lr in lrs:
    for unfreeze in unfreezes:
        for l2_val in l2_vals:
            for fc_size in fc_sizes:
                for batch in batch_sizes:
                    for dropout in dropouts:
                        print('================ ON EN EST AU ' + str(i) + '/108')
                        i+=1
                        if ('batch'+str(batch)+ '-oneFC'+str(fc_size)+'-Dropout'+str(dropout)+'-SGD'+str(lr)+'-unfreeze'+str(unfreeze)+
                        '-l2reg'+str(l2_val)+'.png') not in os.listdir('C:/Users/bbeltzun/Documents/o-o/plots/Molly_good_callbacks/'):
                            training_set = image_dataset_from_directory(train_dir,
                                                                         shuffle=True,
                                                                         batch_size=batch,
                                                                         image_size=(224, 224),
                                                                         label_mode='categorical')
                            val_dataset = image_dataset_from_directory(validation_dir,
                                                                            shuffle=True,
                                                                            batch_size=batch,
                                                                            image_size=(224, 224),
                                                                            label_mode='categorical')
                            base_model = keras.applications.VGG19(
                                weights='imagenet',  
                                input_shape=(224, 224, 3),
                                include_top=False)  

                            base_model.trainable = False
                            if unfreeze != 'none':
                                for layer in base_model.layers[unfreeze:]:
                                    layer.trainable = True

                            inputs = keras.Input(shape=(224, 224, 3))

                            x = data_augmentation(inputs) 

                            x = tf.keras.applications.vgg19.preprocess_input(x)

                            x = base_model(x)
                            x = keras.layers.GlobalAveragePooling2D()(x)
                            x = keras.layers.BatchNormalization()(x)
                            x = keras.layers.Dropout(dropout)(x)
                            x = keras.layers.Dense(fc_size, activation='relu', kernel_regularizer = keras.regularizers.l2(l2=l2_val))(x)
                            outputs = keras.layers.Dense(4, activation = 'softmax')(x)
                            model = keras.Model(inputs, outputs)
                            #for layer in model.layers:
                               # print(layer.name, layer.trainable)
                            model.compile(optimizer=keras.optimizers.SGD(lr), loss= 'categorical_crossentropy',metrics=['accuracy'])
                            
                            early_stop = keras.callbacks.EarlyStopping(
                            monitor = 'val_accuracy',
                            patience = 3,
                            restore_best_weights = True)
                            
                            history = model.fit(training_set, epochs=epoch, validation_data=val_dataset,
                                                batch_size = batch, callbacks = [early_stop])
                            plot_model_history(history)
                            epoch_val_max = np.argmax(history.history['val_accuracy'])
                            val_max = history.history['val_accuracy'][epoch_val_max]
                            with open('C:/Users/bbeltzun/Documents/o-o/plots/Molly_good_callbacks/val_accuracy.txt', 'a') as f:
                                f.writelines('\n'+'batch'+str(batch)+
                                '- oneFC'+str(fc_size)+'- Dropout'+str(dropout)+'- SGD'+str(lr)+'- unfreeze'+str(unfreeze)+
                                '- l2reg'+str(l2_val)+ '- val_accuracy = ' + str(val_max))
                            model.save('C:/Users/bbeltzun/Documents/o-o/plots/Molly_good_callbacks/models/'+
                                       'batch'+str(batch)+ '-oneFC'+str(fc_size)+'-Dropout'+str(dropout)+'-SGD'+str(lr)+'-unfreeze'+str(unfreeze)+
                        '-l2reg'+str(l2_val)+'.h5')
```

```python
predictions = np.array([])
labels = np.array([])
for x,y in val_dataset:
    predictions = np.concatenate([predictions,np.argmax(best_model.predict(x),axis=-1)])
    labels = np.concatenate([labels, np.argmax(y.numpy(), axis=-1)])
confmat = tf.math.confusion_matrix(labels=labels, predictions=predictions).numpy()

print(confmat)
print('accuracy : ' + str(np.diag(confmat).sum()/confmat.sum()))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

confusion_mtx = confusion_matrix(labels, predictions)

cmn = confusion_mtx.astype('float') / confusion_mtx.sum(axis=1)[:, np.newaxis]

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(cmn, annot=True, linewidths=0.01,cmap="Greens",linecolor="gray", fmt= '.1f')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show(block=False)

labels_ticks = ['Autumn','Spring','Summer','Winter']

ax.set_xticklabels(labels_ticks)





# plot the confusion matrix
f,ax = plt.subplots(figsize=(8, 8))
sns.heatmap(confusion_mtx, annot=True, linewidths=0.01,cmap="Greens",linecolor="gray", fmt= '.1f',ax=ax)
ax.set_xticklabels(labels_ticks)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
ax.set_xticklabels(labels_ticks)
ax.set_yticklabels(labels_ticks)
plt.show()
```
